```python
#open the browser

driver_1=webdriver.Chrome()

#load the webpage
driver_1.get("https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile")

time.sleep(5)
```

```python


def apsrtc_route_link(path):
    link_ap = []
    route_ap = []
    
    # Retrieve the route links
    paths = driver_1.find_elements(By.XPATH, path)
    
    for link in paths:
        ap = link.get_attribute("href")
        link_ap.append(ap)
        route_ap.append(link.text)
    
    for url in link_ap:
        driver_1.get(url)
        time.sleep(5)  # Allow more time for the page to load

        try:
            # Initialize WebDriverWait
            wait = WebDriverWait(driver_1, 10)
            
            try:
                # Wait for the pagination element to be present
                pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
                
                # If pagination exists, handle pagination logic
                while True:
                    try:
                        # Find the next button element within the pagination
                        next_button = pagination.find_element(By.XPATH, './/div[@class="DC_117_pageTabs" and text()="Next"]')  # Update the text to match the button text
                        
                        # Scroll to the next button element
                        driver_1.execute_script("arguments[0].scrollIntoView(true);", next_button)
                        
                        # Click the next button
                        next_button.click()
                        
                        # Wait for the next page to load
                        time.sleep(3)
                    
                    except NoSuchElementException:
                        # Exit the loop if no more pages are available
                        break
            except TimeoutException:
                # Handle the case where pagination is not found
                print(f"No pagination element found for URL: {url}")

        except ElementClickInterceptedException as e:
            print(f"ElementClickInterceptedException: {e}")
        except NoSuchElementException as e:
            print(f"NoSuchElementException: {e}")
            continue
    
    return link_ap, route_ap
# Example usage
link_ap, route_ap = apsrtc_route_link("//a[@class='route']")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time

driver_1 = webdriver.Chrome()
Bus_name_ap = []
Bus_type_ap = []
Departure_time_ap = []
Arival_time_ap = []
Rating_ap = []
Duration_ap = []
Prices_ap = []
Seats_Available_ap = []
Route_names = []
Route_links = []

for i, r in df_1.iterrows():
    link = r["route_link"]
    routes = r["route_name"]

    # Loop through each link
    driver_1.get(link)
    time.sleep(2)

    # Click on elements to reveal bus details
    elements = driver_1.find_elements(By.XPATH, f"//a[contains(@href, '{link}')]")
    for element in elements:
        element.click()
        time.sleep(2)

    # Click elements to view bus
    try:
        clicks = driver_1.find_element(By.XPATH, "//div[@class='button']")
        clicks.click()
    except:
        continue
    time.sleep(2)

    scrolling = True
    while scrolling:
        old_page_source = driver_1.page_source
        
        # Use ActionChains to perform a PAGE_DOWN
        ActionChains(driver_1).send_keys(Keys.PAGE_DOWN).perform()
        
        time.sleep(5)  # Adjust sleep time as needed
        
        new_page_source = driver_1.page_source
        
        if new_page_source == old_page_source:
            scrolling = False

    # Extract bus details
    bus_names = driver_1.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
    bus_types = driver_1.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    departure_times = driver_1.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
    arrival_times = driver_1.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
    durations = driver_1.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
    prices = driver_1.find_elements(By.XPATH, '//*[@class="fare d-block"]')
    seats = driver_1.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")
    try:
        ratings = driver_1.find_elements(By.XPATH, "//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    except:
        ratings = []

    # Ensure all lists are the same length
    length = len(bus_names)
    
    for index in range(length):
        Bus_name_ap.append(bus_names[index].text if index < len(bus_names) else '')
        Bus_type_ap.append(bus_types[index].text if index < len(bus_types) else '')
        Departure_time_ap.append(departure_times[index].text if index < len(departure_times) else '')
        Arival_time_ap.append(arrival_times[index].text if index < len(arrival_times) else '')
        Duration_ap.append(durations[index].text if index < len(durations) else '')
        Prices_ap.append(prices[index].text if index < len(prices) else '')
        Seats_Available_ap.append(seats[index].text if index < len(seats) else '')
        Rating_ap.append(ratings[index].text if index < len(ratings) else '')
        Route_links.append(link)
        Route_names.append(routes)
        
print("All bus details of apsrtc are obtained")

# Optionally, you might want to check lengths to ensure they are equal
print(len(Bus_name_ap))
print(len(Bus_type_ap))
print(len(Departure_time_ap))
print(len(Arival_time_ap))
print(len(Duration_ap))
print(len(Prices_ap))
print(len(Seats_Available_ap))
print(len(Rating_ap))
print(len(Route_links))
print(len(Route_names))
```
